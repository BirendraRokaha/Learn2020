### Filtering the data

```python
get_req = (data_f['ConvertedComp'] >= 70000)

data_f.loc[get_req,['Country','YearsCode']]
```

```python
data_f.loc[get_req,'DevType'].value_counts()
```

```python
# Filtering by a list

countries = ['Australia','United Kingdom','Germany','United States','Nepal','India']

filt = data_f['Country'].isin(countries)

data_f.loc[filt,'Country'].value_counts()
```

```python
# Advanced filterring
filt = ((data_f['ConvertedComp'] >= 80000) & (data_f['LanguageWorkedWith'].str.contains('Bash',na =False))) 

data_f.loc[filt,['ConvertedComp','LanguageWorkedWith']]
```

```python
# Count values
data_f.loc[filt,'OpSys'].value_counts()
```
