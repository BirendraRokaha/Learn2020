```python
import sys                             # system module 
import numpy as np                     # number package
import pandas as pd                    # data package
from matplotlib import pyplot as plt   # graphics package
import datetime as dt                  # date and time module

# check versions (overkill, but why not?)
print(f'Python version: {sys.version}')
print(f'Numpy version: {np.__version__}')
print(f'Pandas version: {pd.__version__}')
# print(f'Matplotlib version: {plt.__version__}'
print(f'Today: {dt.date.today()}' )
```

```python
# Get Updated COVID Data file

import requests

# The Url for data
url = 'https://covid.ourworldindata.org/data/owid-covid-data.csv'

# Send a request
file = requests.get(url, allow_redirects=True)

# Save the recived file
with open('covid_data.csv', 'wb') as f:
    f.write(file.content)    
```

```python
data_file = pd.read_csv('covid_data.csv')
data_file
```

```python
nepal = data_file[data_file.location == 'Nepal']

aus = data_file[data_file.location == 'Australia']

ind = data_file[data_file.location == 'India']
```

```python

#Plot Graph for a Country


def plotcoutry(name):    
    c_name = data_file[data_file.location == name]
    
    plt.figure(figsize=(20,10))
    # Getting the data to plot
    
    plt.plot(c_name.date,c_name.total_cases)
    plt.plot(c_name.date,c_name.new_cases)
    
    # labelling
    plt.title(f"Covid chart {name}")
    plt.xlabel("Date")
    plt.ylabel("Number")
    plt.legend(['Total','New'])
    # Display
    plt.show()
    
plotcoutry('Australia')
```

```python
# Compare Total cases of Two countries

def total_compare(name1,name2):    
    a_name = data_file[data_file.location.str.lower() == name1.lower()]
    b_name = data_file[data_file.location.str.lower() == name2.lower()]
    
    plt.figure(figsize=(20,10))
    # Getting the data to plot
    
    plt.plot(a_name.date,a_name.total_cases)
    plt.plot(b_name.date,b_name.total_cases)
    
    # labelling
    plt.title(f"Covid chart {name1} and {name2}")
    plt.xlabel("Date")
    plt.ylabel("Number")
    plt.legend([name1,name2])
    # Display
    plt.show()
    
    
name1 = input('Enter first country: ')
name2 = input('Enter second country: ')

total_compare(name1,name2)
```

```python
# Ploting the graph


plt.figure(figsize=(20,10))

# Getting the data to plot
plt.plot(nepal.date,nepal.total_cases)
plt.plot(nepal.date,nepal.new_cases)


# labelling
plt.title("Covid Nepal")
plt.xlabel("Date")
plt.ylabel("Number")
plt.legend(['Total','New'])

# Display
plt.show()
```

```python
# Ploting the graph


plt.figure(figsize=(20,10))

# Getting the data to plot
plt.plot(aus.date,aus.total_cases)
plt.plot(aus.date,aus.new_cases)


# labelling
plt.title("Covid Australia")
plt.xlabel("Date")
plt.ylabel("Number")
plt.legend(['Total','New'])

# Display
plt.show()
```

```python
# Ploting the graph


plt.figure(figsize=(20,10))

# Getting the data to plot
plt.plot(aus.date,aus.total_cases)
plt.plot(nepal.date,nepal.total_cases)

# labelling
plt.title("Total COIVD cases")
plt.xlabel("Date")
plt.ylabel("Number")
plt.legend(['Aus','Nep'])

# Display
plt.show()
```

```python
# Ploting the graph


plt.figure(figsize=(20,10))
plt.grid(color='#95a5a6', linestyle='--', linewidth=2, axis='y', alpha=0.7)

# Getting the data to plot
plt.plot(aus.date,aus.new_cases)
plt.plot(nepal.date,nepal.new_cases)
plt.plot(nepal.date,ind.new_cases)


# labelling
plt.title("New COIVD cases")
plt.xlabel("Date")
plt.ylabel("Number")
plt.legend(['Aus','Nep','India'])

# Display
plt.show()
```
