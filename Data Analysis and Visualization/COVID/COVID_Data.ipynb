# Working with COVID data

```python
import sys                             # system module 
import numpy as np                     # number package
import pandas as pd                    # data package
from matplotlib import pyplot as plt   # graphics package
import datetime as dt                  # date and time module

# check versions (overkill, but why not?)
print(f'Python version: {sys.version}')
print(f'Numpy version: {np.__version__}')
print(f'Pandas version: {pd.__version__}')
# print(f'Matplotlib version: {plt.__version__}'
print(f'Today: {dt.date.today()}' )
```

```python
data_file = pd.read_csv('Covid_data.csv')
data_file
```

```python
nepal = data_file[data_file.location == 'Nepal']


aus = data_file[data_file.location == 'Australia']
```

```python
# Ploting the graph


plt.figure(figsize=(20,10))

# Getting the data to plot
plt.plot(nepal.date,nepal.total_cases)
plt.plot(nepal.date,nepal.new_cases)


# labelling
plt.title("Covid Nepal")
plt.xlabel("Date")
plt.ylabel("Number")
plt.legend(['Total','New'])

# Display
plt.show()
```

```python
# Ploting the graph


plt.figure(figsize=(20,10))

# Getting the data to plot
plt.plot(aus.date,aus.total_cases)
plt.plot(aus.date,aus.new_cases)

# labelling
plt.title("Covid Australia")
plt.xlabel("Date")
plt.ylabel("Number")
plt.legend(['Total','New'])

# Display
plt.show()
```

```python
# Ploting the graph


plt.figure(figsize=(20,10))

# Getting the data to plot
plt.plot(aus.date,aus.total_cases)
plt.plot(nepal.date,nepal.total_cases)

# labelling
plt.title("Total COIVD cases")
plt.xlabel("Date")
plt.ylabel("Number")
plt.legend(['Aus','Nep'])

# Display
plt.show()
```

```python
# Ploting the graph


plt.figure(figsize=(20,10))
plt.grid(color='#95a5a6', linestyle='--', linewidth=2, axis='y', alpha=0.7)

# Getting the data to plot
plt.bar(aus.date,aus.new_cases)
plt.bar(nepal.date,nepal.new_cases)

# labelling
plt.title("New COIVD cases")
plt.xlabel("Date")
plt.ylabel("Number")
plt.legend(['Aus','Nep'])

# Display
plt.show()
```

```python
# data to plot
n_groups = 4
means_frank = (90, 55, 40, 65)
means_guido = (85, 62, 54, 20)

# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8

rects1 = plt.bar(index, means_frank, bar_width,
alpha=opacity,
color='b',
label='Frank')

rects2 = plt.bar(index + bar_width, means_guido, bar_width,
alpha=opacity,
color='g',
label='Guido')

plt.xlabel('Person')
plt.ylabel('Scores')
plt.title('Scores by person')
plt.xticks(index + bar_width, ('A', 'B', 'C', 'D'))
plt.legend()

plt.tight_layout()
plt.show()
```
