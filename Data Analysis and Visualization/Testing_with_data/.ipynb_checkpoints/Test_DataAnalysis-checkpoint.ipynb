```python
# Loding a csv with Pandas and display the dataset

data_file = pd.read_csv('countries.csv')
```

```python
# Getting isolated data

nepal = data_file[data_file.country == 'Nepal']
aus = data_file[data_file.country == 'Australia']


india = data_file[data_file.country == 'India']
china = data_file[data_file.country == 'China']
```

```python
data_file.year
```

```python
# Ploting the graph

# Getting the data to plot
plt.plot(nepal.year,nepal.population)
plt.plot(aus.year,aus.population)


# labelling
plt.title("Population of Nepal and Australia")
plt.xlabel("years")
plt.ylabel("population")
plt.legend(['Nepal','Australia'])

# Display
plt.show()
```

```python
# Ploting the graph

# Getting the data to plot
# /10 ** 6 to get value in mil
plt.plot(india.year,india.population / 10 ** 6)
plt.plot(china.year,china.population / 10 ** 6)


# labelling
plt.title("India and China")
plt.xlabel("years")
plt.ylabel("population")
plt.legend(['India','China'])

# Display
plt.show()
```

```python
# Getting the Vlues in Percentage form 

pop_ind = ((india.population / india.population.iloc[0]) * 100) - 100

pop_china = ((china.population / china.population.iloc[0]) * 100) - 100

pop_nep = ((nepal.population / nepal.population.iloc[0]) * 100) - 100

pop_aus = ((aus.population / aus.population.iloc[0]) * 100) - 100
```

```python
# Ploting the graph

# Changing the size of the graph
plt.figure(figsize=(20,10))


# Getting the data to plot
plt.plot(india.year,pop_ind,)
plt.plot(china.year,pop_china, '--' )
plt.plot(nepal.year,pop_nep )
plt.plot(aus.year,pop_aus )

# labelling
plt.title(" Population Growth in percentage since 1950")
plt.xlabel("years")
plt.ylabel("percent increase")
plt.legend(['India','China','Nepal','Australia'])



# Display
plt.show()
```

```python
n_groups = nepal.year 
means_frank = pop_nep
means_guido = pop_aus



# # create plot

# fig, ax = plt.subplots()
index = n_groups
bar_width = 0.10
opacity = 0.8
plt.figure(figsize=(15,15))

# print(index)

plt.bar(index, means_frank, bar_width,
alpha=opacity,
color='b',
label='Nepal')

# plt.bar(index , means_guido, bar_width,
# alpha=opacity,
# color='g',
# label='Australia')

plt.xlabel('Year')
plt.ylabel('Growth')
plt.title('Percent Growth')
# plt.xticks(index + bar_width,nepal.year)
plt.legend()

# plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
# Set the axes through the origin
for spine in ['left', 'bottom']:
    ax.spines[spine].set_position('zero')
for spine in ['right', 'top']:
    ax.spines[spine].set_color('none')

ax.set(xlim=(-5, 5), ylim=(-5, 5))
ax.grid()
vecs = ((2, 4), (-3, 3), (-4, -3.5))
for v in vecs:
    ax.annotate('', xy=v, xytext=(0, 0),
                arrowprops=dict(facecolor='blue',
                shrink=0,
                alpha=0.7,
                width=0.5))
    ax.text(1.1 * v[0], 1.1 * v[1], str(v))
plt.show()
```
